```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv(r'C:\Users\hp\OneDrive\Desktop\Python Projects\movies.csv')
```

```python
df.head()
```

```python
#We need to check if there is any missing values in each column so a FOR loop to go through each value in the column 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())  # We're using isnull function returns BOOL type (1's and 0's) and then mean function to give us a percenatage
    print('{} - {}%'.format(col, round(pct_missing *100)))
```

```python
#Remove rows with missing values to elimnate all data disrepcency
df = df.dropna()
```

```python
# Exploring Data Types for our columns
df.dtypes
```

```python
#Cleaning data for selected columns. Changing their data types
df['gross'] = df['gross'].astype('int64')
df['budget'] = df['budget'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
#We want to change default settings to display all the rows when commanded
pd.set_option('display.max_rows', None)
```

```python
#Create a new column with corrected year from Released date
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int) #we used the extract function to take out a pattern of 4 digit number from the string
df.head()
```

```python
#We want to sort the dataset by gross from highest to lowest
df.sort_values(by=['gross'], inplace = False, ascending = False)
df.head()
```

```python
#Remove any duplicates
df.drop_duplicates()
df.head()
```

```python
#Let's try to get a scatter plot to see the correlation between budget and gross

plt.scatter(x=df['budget'], y= df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earning')

plt.ylabel('Budget for Film')

plt.show()
```

```python
#We see that the visual is informative however we want one that displays the Linear Regression to help us understand more
sns.regplot(x='budget', y = 'gross', data=df, scatter_kws={"color": "red"}, line_kws={"color" : "blue"})
plt.show()
```

```python
#Alright, let's have a look at Correlation between other columns 
df.corr(numeric_only=True) #PLEASE NOTE that if not assigned correlation method is autoamtically Pearson instead of Kendall or Spearman Statistcal method
```

```python
# Now Let's try to visualise the correlation scores to make it easier on the eye for information

correlation_matrix= df.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot =True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Budget for Films')

plt.show()
```

```python
# Numerize non-numeric columns in order to add them to Correlation Matrix.
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Movie Dataset')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
# This is to change multi-index Series or DataFrame by moving the inner index level to the columns
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# This is to sort the pairs in a more systematic way
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# This is to find the highest correlation from the pairs
high_corr = sorted_pairs[(sorted_pairs)>0.5]

high_corr
```
